## Iterate and merge train folder

### Subtask:
If the source 'train' folder exists, iterate through its contents (category subfolders and files) and move them to the destination `CombinedDataset/train` folder. Handle potential naming conflicts if necessary (e.g., by skipping or overwriting).


**Reasoning**:
Check if the source train folder exists and if so, iterate through its contents to move them to the destination train folder.



```python
import shutil

source_train_path = os.path.join(source_combined_dataset_path, 'train')
destination_train_path = os.path.join(destination_combined_dataset_path, 'train')

if not os.path.exists(source_train_path):
    print(f"Source train folder not found at {source_train_path}. Skipping train folder merge.")
else:
    print(f"Source train folder found at {source_train_path}. Merging its contents to {destination_train_path}.")
    os.makedirs(destination_train_path, exist_ok=True) # Ensure destination train folder exists

    items_to_process = os.listdir(source_train_path)
    processed_count = 0
    warnings_count = 0

    for item in items_to_process:
        source_item_path = os.path.join(source_train_path, item)
        destination_item_path = os.path.join(destination_train_path, item)

        try:
            if os.path.isdir(source_item_path):
                # If it's a directory, create it in the destination if it doesn't exist
                os.makedirs(destination_item_path, exist_ok=True)
                print(f"Created directory (or already exists): {destination_item_path}")
            else:
                # If it's a file, move it
                shutil.move(source_item_path, destination_item_path)
                # print(f"Moved file: {source_item_path} to {destination_item_path}") # Optional: print each file move
            processed_count += 1
        except FileNotFoundError:
            print(f"Warning: Item not found at {source_item_path}. Skipping.")
            warnings_count += 1
        except Exception as e:
            print(f"Error processing {item}: {e}")
            warnings_count += 1

    print("\n--- Train Folder Merge Summary ---")
    print(f"Source: {source_train_path}")
    print(f"Destination: {destination_train_path}")
    print(f"Items processed (files and folders): {processed_count}")
    print(f"Warnings/Errors encountered: {warnings_count}")
    print("----------------------------------")
```

**Reasoning**:
The subtask of merging the train folder is complete. The next logical step according to the overall task is to merge the valid folder from the source to the destination.



```python
source_valid_path = os.path.join(source_combined_dataset_path, 'valid')
destination_valid_path = os.path.join(destination_combined_dataset_path, 'valid')

if not os.path.exists(source_valid_path):
    print(f"Source valid folder not found at {source_valid_path}. Skipping valid folder merge.")
else:
    print(f"Source valid folder found at {source_valid_path}. Merging its contents to {destination_valid_path}.")
    os.makedirs(destination_valid_path, exist_ok=True) # Ensure destination valid folder exists

    items_to_process = os.listdir(source_valid_path)
    processed_count = 0
    warnings_count = 0

    for item in items_to_process:
        source_item_path = os.path.join(source_valid_path, item)
        destination_item_path = os.path.join(destination_valid_path, item)

        try:
            if os.path.isdir(source_item_path):
                # If it's a directory, create it in the destination if it doesn't exist
                os.makedirs(destination_item_path, exist_ok=True)
                print(f"Created directory (or already exists): {destination_item_path}")
            else:
                # If it's a file, move it
                shutil.move(source_item_path, destination_item_path)
                # print(f"Moved file: {source_item_path} to {destination_item_path}") # Optional: print each file move
            processed_count += 1
        except FileNotFoundError:
            print(f"Warning: Item not found at {source_item_path}. Skipping.")
            warnings_count += 1
        except Exception as e:
            print(f"Error processing {item}: {e}")
            warnings_count += 1

    print("\n--- Valid Folder Merge Summary ---")
    print(f"Source: {source_valid_path}")
    print(f"Destination: {destination_valid_path}")
    print(f"Items processed (files and folders): {processed_count}")
    print(f"Warnings/Errors encountered: {warnings_count}")
    print("----------------------------------")
```

## Set up the data loading and preprocessing pipeline

Define image dimensions, batch size, and functions for loading, preprocessing (resizing, normalization), and augmentation (for training data). Then build the `tf.data` pipelines.

```python
import tensorflow as tf
import tensorflow_datasets as tfds # Using tfds for potential future use or structure examples
import numpy as np

# Define image dimensions and batch size (can be adjusted)
IMG_WIDTH = 224
IMG_HEIGHT = 224
BATCH_SIZE = 32 # Can be adjusted based on your GPU memory

# List of all image file paths for training and validation
# This assumes your data is organized in train/category and valid/category structure
def get_image_paths(base_dir):
    all_image_paths = []
    if os.path.exists(base_dir):
        # Walk through the directory structure
        for root, _, files in os.walk(base_dir):
            for file in files:
                if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
                    all_image_paths.append(os.path.join(root, file))
    return all_image_paths

train_image_paths = get_image_paths(combined_dataset_train_path)
valid_image_paths = get_image_paths(combined_dataset_valid_path)

print(f"Found {len(train_image_paths)} training image files.")
print(f"Found {len(valid_image_paths)} validation image files.")


# Get the unique class names from the training file paths (based on parent directory names)
class_names = sorted(list(set([os.path.basename(os.path.dirname(path)) for path in train_image_paths])))
print(f"Found {len(class_names)} unique classes: {class_names}")

# Create a mapping from class names to integer labels
class_to_label = {name: i for i, name in enumerate(class_names)}
print(f"Class to label mapping: {class_to_label}")


# Function to load image and label
def load_image_and_label_int(file_path):
    """Loads an image and extracts its label from the file path and maps to integer."""
    # Convert the file_path tensor to a Python string using tf.py_function
    # Use the file_path directly as a tensor and decode if necessary within py_function
    file_path_string = tf.py_function(lambda x: x.numpy().decode('utf-8') if isinstance(x.numpy(), bytes) else x.numpy(), [file_path], tf.string)

    # Extract the label string using Python os.path operations within tf.py_function
    label_string = tf.py_function(lambda x: os.path.basename(os.path.dirname(x.numpy().decode('utf-8'))), [file_path], tf.string)

    # Use tf.py_function to wrap the Python dictionary lookup for label mapping
    label_int = tf.py_function(lambda l: class_to_label.get(l.numpy().decode('utf-8'), -1), [label_string], tf.int64) # Use .get with default -1 for safety

    # Read the image file
    img = tf.io.read_file(file_path)

    # Decode the image, ensuring 3 channels for RGB
    try:
        img = tf.image.decode_image(img, channels=3, expand_animations=False) # expand_animations=False for GIF
    except tf.errors.InvalidArgumentError as e:
        # Handle decoding errors - print a warning and return placeholder or skip
        tf.print(f"Warning: Could not decode image {file_path_string}. Error: {e}")
        # Return dummy data or handle later in the pipeline
        # Returning zeros and a dummy label (-1) for problematic images.
        return tf.zeros(shape=(IMG_HEIGHT, IMG_WIDTH, 3), dtype=tf.float32), tf.constant(-1, dtype=tf.int64)


    # Ensure the shapes are defined (required for some operations)
    # img.set_shape([None, None, 3]) # Shape will be determined during batching and resizing
    label_int.set_shape([])


    return img, label_int

# Function for preprocessing (resize and normalize)
def preprocess_image(image, label):
    """Resizes and normalizes images."""
    # Ensure image is float32 for normalization
    image = tf.image.convert_image_dtype(image, dtype=tf.float32)

    # Resize image
    image = tf.image.resize(image, size=(IMG_HEIGHT, IMG_WIDTH))

    # Normalize image (e.g., to range [0, 1])
    # MobileNetV2 typically expects input in the range [-1, 1] if using its preprocessing function
    # or values scaled by 1./255. Let's scale to [-1, 1] for common MobileNetV2 usage.
    # Assumes input is in [0, 1] after convert_image_dtype
    image = (image * 2.0) - 1.0


    return image, label

# Function for data augmentation (apply only to training data)
def augment_image(image, label):
    """ Applies data augmentation to images. """
    # Apply random flip
    image = tf.image.random_flip_left_right(image)

    # Apply random rotation (example)
    # image = tf.image.rot90(image, tf.random.uniform(shape=[], minval=0, maxval=4, dtype=tf.int32))

    # Apply random zoom or crop (example)
    # image = tf.image.random_crop(image, size=[int(IMG_HEIGHT*0.8), int(IMG_WIDTH*0.8), 3])
    # image = tf.image.resize(image, size=(IMG_HEIGHT, IMG_WIDTH))


    # Note: Many other augmentations are available in tf.image


    return image, label

# Filter out problematic images (with label -1)
def filter_problematic_images(image, label):
    return label != -1


# Create a tf.data.Dataset for the training data
train_ds = tf.data.Dataset.from_tensor_slices(train_image_paths)
print("Training dataset created from file paths.")

# Create a tf.data.Dataset for the validation data
valid_ds = tf.data.Dataset.from_tensor_slices(valid_image_paths)
print("Validation dataset created from file paths.")

# Apply the load_image_and_label function to both datasets
# Filter out problematic images *after* loading and attempting to decode
train_ds = train_ds.map(load_image_and_label_int, num_parallel_calls=tf.data.AUTOTUNE).filter(filter_problematic_images)
valid_ds = valid_ds.map(load_image_and_label_int, num_parallel_calls=tf.data.AUTOTUNE).filter(filter_problematic_images)
print("Applied load_image_and_label_int and filtered problematic images to both datasets.")


# Apply the preprocess_image function to both datasets
train_ds = train_ds.map(preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
valid_ds = valid_ds.map(preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
print("Applied preprocess_image to both datasets.")

# Apply the augment_image function to the training dataset
train_ds = train_ds.map(augment_image, num_parallel_calls=tf.data.AUTOTUNE)
print("Applied augment_image to training dataset.")


# Configure datasets for training
# Apply .repeat() BEFORE .shuffle() for better shuffling across epochs
train_buffer_size = len(train_image_paths) if train_image_paths else 1 # Ensure buffer size is at least 1 if list is empty
train_ds = train_ds.repeat().shuffle(buffer_size=train_buffer_size).batch(BATCH_SIZE).prefetch(buffer_size=tf.data.AUTOTUNE)
# Validation dataset does not need shuffling or repeating
valid_ds = valid_ds.batch(BATCH_SIZE).prefetch(buffer_size=tf.data.AUTOTUNE)


print("\nData pipelines configured for training (with repeat for train_ds).")
```

```python
import os
import shutil

# Define the current location of the processed acne train images
source_acne_train_path = os.path.join('/content', 'CombinedDataset', 'train', 'acne')

# Define the correct final destination for the acne train images
destination_acne_train_path = os.path.join('/content', 'unzipped_data', 'CombinedDataset', 'train', 'acne')

# Check if the source acne train folder exists and has contents
if os.path.exists(source_acne_train_path) and os.path.isdir(source_acne_train_path):
    acne_train_images_to_move = [f for f in os.listdir(source_acne_train_path) if os.path.isfile(os.path.join(source_acne_train_path, f))]

    if acne_train_images_to_move:
        print(f"Found {len(acne_train_images_to_move)} acne train images at {source_acne_train_path}.")
        print(f"Merging acne train images to {destination_acne_train_path}.")

        # Ensure the destination directory exists
        os.makedirs(destination_acne_train_path, exist_ok=True)
        print(f"Ensured destination acne train folder exists: {destination_acne_train_path}")

        moved_count = 0
        warnings_count = 0

        # Move each image file
        for image_filename in acne_train_images_to_move:
            source_image_path = os.path.join(source_acne_train_path, image_filename)
            destination_image_path = os.path.join(destination_acne_train_path, image_filename)

            try:
                # Move the file
                shutil.move(source_image_path, destination_image_path)
                moved_count += 1
                # print(f"Moved {image_filename}") # Optional: print each file move
            except FileNotFoundError:
                print(f"Warning: File not found at {source_image_path}. Skipping.")
                warnings_count += 1
            except Exception as e:
                print(f"Error moving {image_filename}: {e}")
                warnings_count += 1

        print("\n--- Acne Train Folder Merge Summary ---")
        print(f"Source: {source_acne_train_path}")
        print(f"Destination: {destination_acne_train_path}")
        print(f"Images successfully moved: {moved_count}")
        print(f"Warnings/Errors encountered: {warnings_count}")
        print("---------------------------------------")

    else:
        print(f"Source acne train folder {source_acne_train_path} is empty. No images to merge.")
else:
    print(f"Source acne train folder not found or is not a directory at {source_acne_train_path}. Skipping merge.")
```
